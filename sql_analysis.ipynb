```python
import pandas as pd
from pandasql import sqldf
from pandasql import load_births
births = load_births()
print(sqldf("SELECT * FROM births where births > 250000 limit 5;", locals()))
```

```python
q = """
        select
            date(date) as DOB,
            sum(births) as "Total Births"
        from
            births
        group by
            date
            limit 10;
"""
# locals vs globals. use globals in function to hardcode it.
print(sqldf(q, locals()))
```

```python
def pysqldf(q):
    return sqldf(q, globals())
```

```python
print(pysqldf(q))
```
