# Matplotlib and Plotting

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 20) #automatically numpy array (min, max, samples)
y = 2*x**2 #just some random thing 2x^2

plt.plot(x,y)
plt.show()
```

```python
#customizations
plt.plot(x,y,color='red',marker='o', linestyle='--',label='my plot')

#setting the x and y limits
plt.xlim(-2,12)
plt.ylim(-25,225)

#grid
plt.grid()

#legend with label
plt.legend()

#show figure
plt.show()
```

```python
#bar chart
categories = ['red','blue','black','green']
values=[10,2,5,7]

plt.bar(categories,values)
plt.title('fav colors')
plt.xlabel('types of colors')
plt.ylabel('Ppl who liked')
plt.show()
```

```python
mean=80
stdev=5
rand_samp=1000
normal_dist=np.random.normal(mean,stdev,rand_samp)

plt.hist(normal_dist,11) #11 is the bin
plt.show()
```

```python
#gives control over the figure object (background window) and the axis object (the plot itself)
fig, ax = plt.subplots()

#add data to axis object
ax.plot(x,y,label='frustration curve')

#set attributes on axis object
ax.set_title('the better way to plot')
ax.set_xlabel('time spent working on an assignmnet')
ax.set_ylabel('frustration with assignment')
 
ax.set_xlim(-2,12)
ax.set_ylim(-25,225)

#you don't set the grid and legend
ax.grid()
ax.legend()

#show figure
fig.show()
```

```python
x=np.linspace(0,10,100)
y=2*x**2
z=3*x+5

#create figure/axis objects
fig, ax = plt.subplots()

#change background colors of fig/ax
fig.set_facecolor('lightgray')
ax.set_facecolor('wheat')

#change the border box that separates the axis object from the figure object
ax.spines['right'].set_color(None) #off
ax.spines['top'].set_color(None)
ax.spines['left'].set_color('maroon')
ax.spines['bottom'].set_color('aqua')

#add data to axis objects
ax.plot(x,y,color='blue',label='curved')
ax.plot(x,z,color='red',label='straight')
ax.legend()

#show figure
fig.show()
```

```python
x=np.linspace(0,10,100)
y=2*x**2
z=3*x+50

#making the figure/axis object
rows = 1 #one row
cols = 2 #with 2 plots in it
fig, ax = plt.subplots(rows,cols,figsize=(10,5)) #make figure wider for extra plot

#change figure color to show the difference between figure and axes
fig.set_facecolor('lightgray')

#customizing axis 0 (the left figure)
ax[0].plot(x,y,color='red',label='curved')
ax[0].set_title('plot 1')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()

#customizing axis 1 (the right figure)
ax[1].plot(x,z,color='purple',label='straight')
ax[1].set_title('plot 2')
ax[1].set_xlabel('different x')
ax[1].set_ylabel('not same y')
ax[1].legend()

#superior (super, upper) title on the figure window
fig.suptitle('top level figure title')

#show the figure
fig.show()
```

```python
#define a quadratic function
def quad(a,b,c,x):
    return a*x**2 + b*x +c

x=np.linspace(-5,5,100)
y1=quad(1,2,3,x)
y2=quad(-3,1,2,x)
y3=quad(-2,3,1,x)
y4=quad(2,1,-3,x)
y5=quad(3,2,1,x)
y6=quad(1,-3,2,x)

#creating the figure, axis objects
rows=2
cols=3
fig, ax = plt.subplots(rows,cols,figsize=(20,10))

#add plots to axis
ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[0,2].plot(x,y3)
ax[1,0].plot(x,y4)
ax[1,1].plot(x,y5)
ax[1,2].plot(x,y6)

#show figure that holds all 6 axes on it
fig.show()
```
